### Step 4 - Apply Preprocessing on X

```python
cat = list(X.columns[X.dtypes == "object"])
con = list(X.columns[X.dtypes != "object"])
```

```python
cat
```

```python
con
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipe = make_pipeline(
    SimpleImputer(strategy = "median"),
    StandardScaler()
)
```

```python
cat_pipe = make_pipeline(
    SimpleImputer(strategy = "most_frequent"),
    OneHotEncoder(handle_unknown = "ignore", sparse_output = False)
)
```

```python
pre = ColumnTransformer(
    [
        ("num", num_pipe, con),
        ("cat", cat_pipe, cat)
    ]
).set_output(transform = "pandas")
```

```python
X_pre = pre.fit_transform(X)
X_pre.head()
```

### Step 5 - Apply train test split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size = 0.2, random_state = 42)
```

### Step 6 - Algorithm Evaluation

SVM, KNN is skipped because it is suitable for smaller datasets upto 1000 rows

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
import xgboost as xgb
```

```python
models = [
    LogisticRegression(),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    GradientBoostingClassifier(),
    xgb.XGBClassifier()
]
```

```python
from sklearn.metrics import f1_score
from sklearn.model_selection import cross_val_score
```

```python
def evaluate_single_model(model, xtrain, ytrain, xtest, ytest):
    # Fit the model
    model.fit(xtrain, ytrain)

    # Predict results on train and test
    ypred_train = model.predict(xtrain)
    ypred_test = model.predict(xtest)

    # Check f1 macro on train and test
    f1_train = f1_score(ytrain, ypred_train, average = "macro")
    f1_test = f1_score(ytest, ypred_test, average = "macro")

    # 5 fold crossval idation on train data
    scores = cross_val_score(model, xtrain, ytrain, cv = 5, scoring = "f1_macro", n_jobs = -1)
    f1_cv = scores.mean()

    # Create the results dictionary
    res = {
        "name":type(model).__name__,
        "model":model,
        "f1_train":f1_train,
        "f1_test":f1_test,
        "f1_cv":f1_cv
    }
    return res
```

```python
def algo_evaluation(models: list, xtrain, ytrain, xtest, ytest):
    # Get results for every model
    results = []
    for model in models:
        res = evaluate_single_model(model, xtrain, ytrain, xtest, ytest)
        print(res)
        results.append(res)

    # Convert the results to dataframe
    df = pd.DataFrame(results)

    # Sort the dataframe by f1_cv
    sort_df = df.sort_values(by = "f1_cv", ascending = False).reset_index(drop = True)

    # Get the best model
    best_model = df.loc[0, "model"]

    return sort_df.round(4), best_model
```
